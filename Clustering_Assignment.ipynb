# Answers to Theoretical Questions on Unsupervised Learning and Clustering

## 1. What is unsupervised learning in the context of machine learning?
Unsupervised learning is a type of machine learning where algorithms learn patterns from untagged data without predefined labels or output variables. The system tries to find hidden structures or patterns in the data on its own.

## 2. How does K-Means clustering algorithm work?
K-Means works by:
1. Randomly initializing K cluster centroids
2. Assigning each data point to the nearest centroid
3. Recalculating centroids as the mean of all points in each cluster
4. Repeating steps 2-3 until centroids stabilize or max iterations reached

## 3. Explain the concept of a dendrogram in hierarchical clustering.
A dendrogram is a tree-like diagram that records the sequences of merges or splits in hierarchical clustering. It shows how clusters are formed at different distances/similarities, with the height representing the distance between merging clusters.

## 4. What is the main difference between K-Means and Hierarchical Clustering?
K-Means requires specifying the number of clusters (K) upfront and partitions data into non-overlapping clusters, while hierarchical clustering builds a hierarchy of clusters (dendrogram) without requiring K upfront and allows exploring clusters at different levels.

## 5. What are the advantages of DBSCAN over K-Means?
DBSCAN advantages:
- Doesn't require specifying number of clusters
- Can find arbitrarily shaped clusters
- Identifies noise/outliers
- Works well with clusters of varying densities
- Not sensitive to initial centroid placement

## 6. When would you use Silhouette Score in clustering?
Silhouette Score is used to:
- Evaluate clustering quality when true labels are unknown
- Compare different clustering results
- Determine optimal number of clusters
- Assess how well each point fits its assigned cluster

## 7. What are the limitations of Hierarchical Clustering?
Limitations include:
- Computationally expensive (O(n³) for most implementations)
- Sensitive to noise and outliers
- Once points are assigned, they can't be moved to different clusters
- Difficult to handle large datasets
- Different linkage methods can produce different results

## 8. Why is feature scaling important in clustering algorithms like K-Means?
Feature scaling is crucial because K-Means uses distance measures (typically Euclidean). Without scaling, features with larger scales dominate the distance calculation, potentially distorting cluster formation. Scaling ensures all features contribute equally.

## 9. How does DBSCAN identify noise points?
DBSCAN identifies noise points as those that:
- Are not within ε distance (eps) of any core point
- Don't have enough neighbors (minPts) to form their own dense region

## 10. Define inertia in the context of K-Means.
Inertia is the sum of squared distances of samples to their nearest cluster center. It measures how internally coherent clusters are, with lower values indicating better clustering.

## 11. What is the elbow method in K-Means clustering?
The elbow method is a technique to determine the optimal number of clusters (K) by:
1. Running K-Means for different K values
2. Plotting inertia against K
3. Looking for the "elbow" point where inertia starts decreasing linearly rather than rapidly

## 12. Describe the concept of "density" in DBSCAN.
In DBSCAN, density is defined by two parameters:
- ε (eps): radius of the neighborhood around a point
- minPts: minimum number of points within ε distance
A region is dense if it contains at least minPts within ε distance of each other.

## 13. Can hierarchical clustering be used on categorical data?
Yes, hierarchical clustering can be used with categorical data by employing appropriate distance metrics like:
- Hamming distance
- Jaccard distance
- Simple matching coefficient
- Gower's distance (for mixed data types)

## 14. What does a negative Silhouette Score indicate?
A negative Silhouette Score indicates that:
- Many points are likely assigned to wrong clusters
- The average distance to points in the same cluster is greater than to points in the nearest neighboring cluster
- The clustering configuration may be worse than random assignment

## 15. Explain the term "linkage criteria" in hierarchical clustering.
Linkage criteria determine how distances between clusters are calculated. Common types:
- Single linkage: minimum distance between clusters
- Complete linkage: maximum distance between clusters
- Average linkage: average distance between all pairs
- Ward's method: minimizes variance when merging clusters

## 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
K-Means performs poorly in these cases because:
- It assumes clusters are spherical and equally sized
- Centroids may be dominated by larger clusters
- Distance-based assignment favors equal-sized clusters
- It can't handle irregular cluster shapes well

## 17. What are the core parameters in DBSCAN, and how do they influence clustering?
Core parameters:
- ε (eps): Radius for neighborhood search
  - Larger ε: fewer, larger clusters
  - Smaller ε: more, smaller clusters
- minPts: Minimum points to form a dense region
  - Higher minPts: more noise points, fewer clusters
  - Lower minPts: more clusters, less noise

## 18. How does K-Means++ improve upon standard K-Means initialization?
K-Means++ improves initialization by:
1. Selecting first centroid randomly
2. Choosing subsequent centroids with probability proportional to squared distance from nearest existing centroid
3. This spreads initial centroids better, reducing poor local optima and convergence time

## 19. What is agglomerative clustering?
Agglomerative clustering is a bottom-up hierarchical clustering approach where:
- Each point starts as its own cluster
- Closest pairs of clusters are merged iteratively
- Continues until all points are in one cluster or stopping condition met

## 20. What makes Silhouette Score a better metric than just inertia for model evaluation?
Silhouette Score is better because:
- Considers both cohesion (within-cluster distance) and separation (between-cluster distance)
- Provides normalized score (-1 to 1) that's comparable across datasets
- Works for any distance metric, not just Euclidean
- Can evaluate individual points' cluster fit
- Not biased towards spherical clusters like inertia is

```python
# Practical Clustering Assignment Solutions
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D

# Set random seed for reproducibility
np.random.seed(42)

## Question 21
print("Question 21: K-Means on synthetic data with 4 centers")
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=200, c='red', marker='X')
plt.title("K-Means Clustering with 4 Centers")
plt.show()

## Question 22
print("\nQuestion 22: Agglomerative Clustering on Iris dataset")
iris = datasets.load_iris()
agg = AgglomerativeClustering(n_clusters=3)
labels = agg.fit_predict(iris.data)

print("First 10 predicted labels:", labels[:10])

## Question 23
print("\nQuestion 23: DBSCAN on moons data with outliers highlighted")
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)
X = StandardScaler().fit_transform(X)
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.title("DBSCAN on Moons Data (Outliers in Yellow)")
plt.show()

## Question 24
print("\nQuestion 24: K-Means on standardized Wine dataset")
wine = datasets.load_wine()
X = StandardScaler().fit_transform(wine.data)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

unique, counts = np.unique(labels, return_counts=True)
print("Cluster sizes:", dict(zip(unique, counts)))

## Question 25
print("\nQuestion 25: DBSCAN on circles data")
X, y = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)
X = StandardScaler().fit_transform(X)
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.title("DBSCAN on Circles Data")
plt.show()

## Question 26
print("\nQuestion 26: K-Means on Breast Cancer data")
cancer = datasets.load_breast_cancer()
X = MinMaxScaler().fit_transform(cancer.data)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

print("Cluster centroids:\n", kmeans.cluster_centers_)

## Question 27
print("\nQuestion 27: DBSCAN on blobs with varying std")
X, y = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)
X = StandardScaler().fit_transform(X)
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.title("DBSCAN on Blobs with Varying STD")
plt.show()

## Question 28
print("\nQuestion 28: Digits clustering with PCA reduction")
digits = datasets.load_digits()
X = PCA(n_components=20).fit_transform(digits.data)
kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='tab20')
plt.title("Digits Clustering (PCA reduced to 20D, plotted first 2 components)")
plt.show()

## Question 29
print("\nQuestion 29: Silhouette scores for k=2 to 5")
X, y = make_blobs(n_samples=500, centers=4, random_state=42)
silhouette_scores = []
k_values = range(2,6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8,6))
plt.bar(range(len(k_values)), silhouette_scores, tick_label=k_values)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for k=2 to 5')
plt.show()

## Question 30
print("\nQuestion 30: Hierarchical clustering dendrogram for Iris")
iris = datasets.load_iris()
X = iris.data

plt.figure(figsize=(12,6))
linkage_matrix = linkage(X, 'average')
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

## Question 41
print("\nQuestion 41: Silhouette score evaluation for KMeans")
X, y = make_blobs(n_samples=500, centers=5, random_state=42)
kmeans = KMeans(n_clusters=5, random_state=42)
labels = kmeans.fit_predict(X)
score = silhouette_score(X, labels)

print(f"Silhouette Score for 5 clusters: {score:.3f}")

## Question 42
print("\nQuestion 42: Agglomerative Clustering on PCA-reduced Breast Cancer data")
cancer = datasets.load_breast_cancer()
X = PCA(n_components=2).fit_transform(cancer.data)
agg = AgglomerativeClustering(n_clusters=2)
labels = agg.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.title("Agglomerative Clustering on Breast Cancer Data (PCA reduced)")
plt.show()

## Question 43
print("\nQuestion 43: Comparing KMeans and DBSCAN on circles data")
X, y = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)
X = StandardScaler().fit_transform(X)

# KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c=kmeans_labels, cmap='viridis')
plt.title("KMeans on Circles Data")

plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,1], c=dbscan_labels, cmap='viridis')
plt.title("DBSCAN on Circles Data")
plt.show()

## Question 44
print("\nQuestion 44: Silhouette plot for Iris KMeans clustering")
iris = datasets.load_iris()
X = iris.data
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)
silhouette_vals = silhouette_samples(X, labels)

plt.figure(figsize=(8,6))
y_lower = 10
for i in range(3):
    cluster_silhouette_vals = silhouette_vals[labels == i]
    cluster_silhouette_vals.sort()
    y_upper = y_lower + cluster_silhouette_vals.shape[0]
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_vals)
    y_lower = y_upper + 10

plt.axvline(x=silhouette_score(X, labels), color="red", linestyle="--")
plt.title("Silhouette Plot for Iris KMeans Clustering")
plt.show()

## Question 45
print("\nQuestion 45: Agglomerative Clustering with average linkage")
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
agg = AgglomerativeClustering(n_clusters=4, linkage='average')
labels = agg.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.title("Agglomerative Clustering with Average Linkage")
plt.show()

## Question 46
print("\nQuestion 46: Wine cluster visualization with pairplot")
wine = datasets.load_wine()
X = StandardScaler().fit_transform(wine.data)
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

df = pd.DataFrame(X[:,:4], columns=wine.feature_names[:4])
df['cluster'] = labels
sns.pairplot(df, hue='cluster', palette='viridis')
plt.suptitle("Wine Cluster Pairplot (First 4 Features)", y=1.02)
plt.show()

## Question 47
print("\nQuestion 47: DBSCAN on noisy blobs")
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)
X = StandardScaler().fit_transform(X)
dbscan = DBSCAN(eps=0.5, min_samples=10)
labels = dbscan.fit_predict(X)

unique, counts = np.unique(labels, return_counts=True)
print("Cluster counts:", dict(zip(unique, counts)))
print("Noise points:", counts[-1] if -1 in labels else 0)

## Question 48
print("\nQuestion 48: Digits clustering with t-SNE reduction")
digits = datasets.load_digits()
X = TSNE(n_components=2).fit_transform(digits.data)
agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X)

plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='tab20')
plt.title("Digits Clustering (t-SNE reduced to 2D)")
plt.show()
```
